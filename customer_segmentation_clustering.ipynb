Lãng phí chi tiêu quảng cáo là vấn đề lớn mà tất cả các doanh nghiệp phải đối mặt. Chỉ riêng trong năm 2021, 3 tỷ đô la Mỹ chi tiêu cho quảng cáo đã bị lãng phí. Một trong những nguyên nhân chính dẫn đến sự lãng phí này là sai lầm khi nhắm mục tiêu sai đối tượng. Mục tiêu của chiến dịch quảng cáo là thu hút sự chú ý của càng nhiều khách hàng càng tốt. Do đó, nó nên được tiếp thị hướng tới nhóm có nhiều khả năng mua sản phẩm nhất để mang lại nhiều doanh thu hơn.

**Mục tiêu dự án**
- Trong dự án này, chúng ta sẽ khám phá dữ liệu được thu thập ở một siêu thị về khách hàng của siêu thị đó. Bộ dữ liệu này cung cấp cho chúng ta một số thông tin về khách hàng (Bao gồm thông tin cơ bản và sở thích mua hàng của họ). Chúng ta sẽ chia khách hàng thành các phân khúc khách nhau bằng các kỹ thuật phân cụm. Tiếp theo đó, Chúng ta sẽ cố gắng đánh giá việc phân cụm bằng một số số liệu phổ biến. Tuy nhiên, chúng ta sẽ phải nghiên cứu sâu về cụm để đánh giá tính đúng đắn của thuật toán phân cụm.

- Vào cuối dự án, chúng ta sẽ có một mô tả thống kê về các phân khúc khách hàng khác nhau và thói quen chi tiêu của họ. Sau đó, siêu thị có thể tận dụng các phân khúc đó giúp tìm ra những khách hàng lý tưởng. Nó còn giúp siêu thị hiểu rõ hơn về khách hàng của mình và giúp họ lập kế hoạch tiếp thị phù hợp dễ dàng hơn, chẳng hạn như sửa đổi và quảng cáo sản phẩm cho các khách hàng khác nhau theo nhu cầu, hành vi và mối quan tâm cụ thể.

```python
def display_missing(df, feature_cols):
    n_rows = df.shape[0]
    for col in feature_cols:
        missing_count = df[col].isnull().sum()
        if missing_count > 0:
            print(f"{col} has {missing_count*100/n_rows:.2f}% missing values.")
            
display_missing(data, feature_cols)
```

Các giá trị bị thiếu của thuộc tính Income không ảnh hưởng đến đáng kể đến bộ dữ liệu. Vì thế cần xóa những giá trị này ra khỏi dữ liệu.

```python
data = data.dropna()
print("The total number of data-points after removing the rows with missing values are:", len(data))
```

Trong bước tiếp theo, ta sẽ tạo một thuộc tính mới từ **"Dt_Customer"** là **("Customer_For")** cho biết số ngày khách hàng bắt đầu mua sắm tại cửa hàng. Cách tính ta lấy ngày đăng ký thành viên của các khách hàng so với ngày đăng ký thành viên của khách hàng gần nhất.

```python
data["Dt_Customer"] = pd.to_datetime(data["Dt_Customer"])
dates = []
for i in data["Dt_Customer"]:
    i = i.date()
    dates.append(i)  
```

```python
#Created a feature "Customer_For"
days = []
d1 = max(dates)
for i in dates:
    delta = (d1 - i).days
    days.append(delta)
data["Customer_For"] = days
data["Customer_For"] = pd.to_numeric(data["Customer_For"], errors="coerce")
```

Bây giờ chúng ta sẽ khám phá các giá trị duy nhất trong các tính năng phân loại để có ý tưởng rõ ràng về dữ liệu.

```python
print("Total categories in the feature Marital_Status:\n", data["Marital_Status"].value_counts(), "\n")
print("Total categories in the feature Education:\n", data["Education"].value_counts())
```

- Thuộc tính **Marital_Status**: Có 8 loại giá giá trị, khi xem xét kĩ gàng có thể gom một số giá trị lại với nhau như: Alone, Single, widow, YOLO, Divorced, Absurb được đại diện bởi giá trị **Alone**; Married, Together bởi **Partner**. 
- Thuộc tính **Education**: có 5 loại giá trị, có thể gom lại thành giá trị mới gồm **Basic** đại diện cho Basic, **Undergraduate** đại diện cho 2n Cycle , **Graduate** đại diện cho Graduation và **Postgraduate** đại diện cho Master, PhD.


**Trong phần tiếp theo, ta sẽ thực hiện các bước sau để thiết kế một số tính năng mới:**
- Tạo thuộc tính **Age** của khách hàng theo **Year_Birth** cho biết năm sinh của người tương ứng tính đến thời điểm gần nhất của năm bộ dữ liệu được thu thập (2015).
- Tạo một thuộc tính khác **Spent** cho biết tổng số tiền mà khách hàng đã chi tiêu trong các danh mục khác nhau trong khoảng thời gian hai năm.
- Tạo một thuộc tính **Children** để cho biết tổng số trẻ em trong một hộ gia đình là trẻ em **(Kidhome)** và thanh thiếu niên **(Teenhome)**. 
- Để hiểu rõ hơn về hộ gia đình, Tạo thuộc tính cho biết **Family_Size**, thuộc tính bằng công thức xem thuộc tính **Marital_Status** với Alone = 1 (nghĩa là sống 1 mình), Partner = 2 (nghĩa là có hai người sống chung một nhà) cộng cho số người con của họ trong thuộc tính Children.
- Tạo thuộc tính **Is_Parent** để hiển thị trạng thái làm cha mẹ, nếu một người có con (thuộc tính Children có giá trị từ 1 trở lên), nghĩa là người này làm cha hoặc mẹ nên giá trị **Is_Parent** sẽ bằng 1, ngược lại bằng 0. 
- Tạo tính năng **Num_Accepted** từ tổng số chiến dịch tiếp thị trước đó đã được khách hàng chấp nhận.


```python
#Feature Engineering
#Age of customer today 
data["Age"] = 2015-data["Year_Birth"]

#Total spendings on various items
data["Spent"] = data["MntWines"] + data["MntFruits"] + data["MntMeatProducts"] + data["MntFishProducts"] + data["MntSweetProducts"] + data["MntGoldProds"]

#Deriving living situation by marital status"Alone"
data["Marital_Status"]=data["Marital_Status"].replace({"Married":"Partner", "Together":"Partner", "Absurd":"Alone", "Widow":"Alone", "YOLO":"Alone", "Divorced":"Alone", "Single":"Alone",})

#Feature indicating total children living in the household
data["Children"]=data["Kidhome"]+data["Teenhome"]

#Feature for total members in the householde
data["Family_Size"] =  data["Marital_Status"].replace({"Alone": 1, "Partner":2}) + data["Children"]

#Feature pertaining parenthood
data["Is_Parent"] = np.where(data.Children> 0, 1, 0)

#Segmenting education levels in three groups
data["Education"]=data["Education"].replace({"Basic":"Basic","2n Cycle":"Undergraduate", "Graduation":"Graduate", "Master":"Postgraduate", "PhD":"Postgraduate"})

#For clarity
#data=data.rename(columns={"MntWines": "Wines","MntFruits":"Fruits","MntMeatProducts":"Meat","MntFishProducts":"Fish","MntSweetProducts":"Sweets","MntGoldProds":"Gold"})

# Clear description of feature
data.rename(columns={'MntWines':'Wines', 'MntFruits':'Fruits', 'MntMeatProducts':'Meats', 'MntFishProducts':'Fish', 'MntSweetProducts':'Sweets', 'MntGoldProds':'Golds'}, inplace=True)
data.rename(columns={'NumWebPurchases':'Web', 'NumCatalogPurchases':'Catalog', 'NumStorePurchases':'Store'}, inplace=True)

#Create 'Num_Accepted' feature from the sum of previous marketting campaigns that were accepted by the customer
data['Num_Accepted'] = data.AcceptedCmp1.values + data.AcceptedCmp2.values + \
                                data.AcceptedCmp3.values + data.AcceptedCmp4.values + data.AcceptedCmp5.values

#Dropping some of the redundant features
to_drop = ["Z_CostContact", "Z_Revenue", "Year_Birth", "ID","Kidhome","Teenhome","Children"]
data = data.drop(to_drop, axis=1)
print("Bây giờ chúng ta có một số thuộc tính mới, hãy xem số liệu thống kê của dữ liệu.")
data.head()
```

Tiến hành tiền xử lý dữ liệu trước khi phân cụm.
Các bước thực hiện:
*	Dùng LabelEncoder đưa các biến ở dạng phân loại (categorical) về dạng số (numerical).
*	Sử dụng StandardScaler để scaling các giá trị của tất cả các feature.
* Tạo dataframe con cho dữ liệu đã được giảm chiều. 

```python
#Lấy danh sách các đối tượng categorical
s = (data.dtypes == 'object')
object_cols = list(s[s].index)

print("Categorical variables in the dataset:", object_cols)
```

```python
#Label Encoding
LE=LabelEncoder()
for i in object_cols:
    data[i]=data[[i]].apply(LE.fit_transform)
    
print("All features are now numerical")
```

```python
#Tạo một bảng sao dữ liệu
ds = data.copy()

# Tạo một tập hợp con của khung dữ liệu bằng cách loại bỏ các thuộc tính trên các AcceptedCmp và Promotion
cols_del = ["AcceptedCmp1", "AcceptedCmp2", "AcceptedCmp3", "AcceptedCmp4", "AcceptedCmp5", "Complain", "Num_Accepted", "Response", "Dt_Customer"]
ds = ds.drop(cols_del, axis=1)

# Scaling
scaler = StandardScaler()
scaler.fit(ds)
scaled_ds = pd.DataFrame(scaler.transform(ds),columns= ds.columns )
print("All features are now scaled")
```

```python
#Scaled data to be used for reducing the dimensionality
print("Dataframe to be used for further modelling:")
scaled_ds.head()
```

# <p style="background-color:#682F2F;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px;">DIMENSIONALITY REDUCTION</p>
Trong vấn đề này, có nhiều nhân tố mà trên cơ sở đó ảnh hưởng đến việc phân cụm. Những nhân tố này cơ bản là thuộc tính hoặc là tính năng. Càng nhiều thuộc tính thì càng khó để làm việc với chúng. Nhiều thuộc tính trong số này là tương quan và do đó dư thừa. Đó là lý do tại sao phải thực hiện giảm chiều dữ liệu với những tính năng được chọn trước khi thực hiện phân cụm.

**Principal component analysis (PCA)** là một thuật toán thống kê sử dụng phép biến đổi trực giao để biến đổi một tập dữ liệu từ một không gian nhiều chiều (không thể biểu diễn trên hệ tọa độ) thành một không gian ít chiều hơn (2 hoặc 3 chiều) nhằm dễ dàng biểu diễn trên hệ trục tọa độ để có một cái nhìn tổng quan về dữ liệu

*Ưu điểm :*

* Giảm số chiều của không gian chứa dữ liệu khi nó có số chiều lớn, không thể thể hiện trong không gian 2 hay 3 chiều.
* Xây dựng những trục tọa độ mới, thay vì giữ lại các trục của không gian cũ, nhưng lại có khả năng biểu diễn dữ liệu tốt tương đương, và đảm bảo độ biến thiên của dữ liệu trên mỗi chiều mới.
* Tạo điều kiện để các liên kết tiềm ẩn của dữ liệu có thể được khám phá trong không gian mới, mà nếu đặt trong không gian cũ thì khó phát hiện vì những liên kết này không thể hiện rõ.
* Đảm bảo các trục tọa độ trong không gian mới luôn trực giao đôi một với nhau, mặc dù trong không gian ban đầu các trục có thể không trực giao.

**Các bước ở phần này:**
* Giảm chiều dữ liệu với PCA.
* Vẽ dataframe đã giảm chiều.

**Vậy chúng ta cần chọn số chiều khi dùng PCA là bao nhiêu ?**

Biểu đồ tương quan giữa tỉ lệ phương sai tích lũy với số chiều dữ liệu:

```python
pca = PCA().fit(ds)

print(pca.explained_variance_ratio_.cumsum())

plt.plot(pca.explained_variance_ratio_.cumsum())
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
```

**Giảm chiều dữ liệu với thuật toán PCA:**

Từ biểu đồ scree, ta có thể thấy rằng, chỉ cần từ 1 thành phần chính trở lên có thể giải thích hơn 99,94% phương sai trong dữ liệu gốc, nghĩa là dữ liệu có mối tương quan cao. Do đó chúng ta có thể giảm số lượng biến xuống từ ít nhất 1 mà vẫn giữ lại hơn 99,94% thông tin có trong dữ liệu gốc.

Chúng ta sẽ thực hiện giảm chiều dữ liệu còn 3. Tại vì dữ liệu 3 chiều sẽ giúp việc vẽ và hình dung chính xác hơn.

```python
#Initiating PCA to reduce dimentions aka features to 3
pca = PCA(n_components=3)
pca.fit(scaled_ds)
PCA_ds = pd.DataFrame(pca.transform(scaled_ds), columns=(["col1","col2", "col3"]))
PCA_ds.describe().T
```

```python
#Hình chiếu 3D của dữ liệu trong kích thước đã giảm
x =PCA_ds["col1"]
y =PCA_ds["col2"]
z =PCA_ds["col3"]
#To plot
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(x,y,z, c="maroon", marker="o" )
ax.set_title("A 3D Projection Of Data In The Reduced Dimension")
plt.show()
```

# <p style="background-color:#682F2F;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px;">CLUSTERING</p>

Sau khi giảm chiều dữ liệu đến 3 chiều, tiến hành phân cụm (cluster) bằng thuật toán Kmeans.

**Các bước phân cụm:**
* Sử dụng đồ thị Elbow để quyết định số cụm k.
* Tiến hành phân cụm bằng thuật toán Kmeans.
* Kiểm tra các cụm được hình thành qua biểu đồ phân tán.

```python
# Quick examination of elbow method to find numbers of clusters to make.
print('Elbow Method to determine the number of clusters to be formed:')
Elbow_M = KElbowVisualizer(KMeans(), k=10)
Elbow_M.fit(PCA_ds)
Elbow_M.show()
```

Để kiểm tra các cụm được hình thành, chúng ta hãy xem phân phối 3-D của các cụm.

Chúng ta sẽ sử dụng k = 4, do đó sẽ có bốn cụm trong tập dữ liệu của chúng ta.
Chúng ta sẽ sử dụng Thuật toán KMeans để phân cụm.

```python
km = KMeans(n_clusters = 4,random_state = 42)

y_km = km.fit_predict(PCA_ds)
PCA_ds["Clusters"] = y_km
data["Clusters"]= y_km
```

```python
#Plotting the clusters
fig = plt.figure(figsize=(10,8))
ax = plt.subplot(111, projection='3d', label="bla")
ax.scatter(x, y, z, s=40, c=PCA_ds["Clusters"], marker='o', cmap = 'RdBu_r' )
ax.set_title("The Plot Of The Clusters")
plt.show()
```

```python
y_reduced = km.fit(PCA_ds)
clusters = y_reduced.labels_

# add cluster labels to the original dataset
data['Cluster'] = clusters
data.head()
```
